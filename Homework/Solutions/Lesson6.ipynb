```
#Numpy Arrays

import numpy as np
no_of_seats_per_class = [100, 80, 60, 90]  # list of seats available per class
no_of_students_per_class = [82, 34, 49, 88]  # list of students attended per class

# find no. of seats vacant per class using numpy array operations
# START HERE
np_seats = np.array(no_of_seats_per_class)
np_students = np.array(no_of_students_per_class)
print("No. of seats vacant per class :",np_seats - np_students)

# Expected result:
# No. of seats vacant per class : [18 46 11  2]
```

```
# using array of booleans 

import numpy as np
list_of_class = ["Politics", "Engineering", "Biology", "Maths"]
no_of_seats_per_class = [100, 80, 60, 90]  # list of seats available per class
no_of_students_per_class = [82, 80, 49, 90]  # list of students attended per class

# print class names with full attendance using numpy array operations and array of booleans:
# START HERE

#Converting arrays to numpy arrays
np_list_of_class = np.array(list_of_class)
np_no_of_seats_per_class = np.array(no_of_seats_per_class)
np_no_of_students_per_class = np.array(no_of_students_per_class)

#Finding class names with full attendance by slicing using numpy array of booleans
print(np_list_of_class[np_no_of_seats_per_class == np_no_of_students_per_class])

# Expected result:
# ['Engineering' 'Maths']
```
