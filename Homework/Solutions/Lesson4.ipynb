<span>1. </span> Referring to the last code block of Lesson 4, what if we just want to know the team with the most number of steps? Write code / modify the code to obtain output like this:
```
The winning team is XXX, with xxx steps.
```

````python
# Start question 1 here.
# Print total steps by each team
import csv
def total_steps_per_person(participant_data, header): # each person's data is in one row
    total_steps = 0
    for col_index, fieldNames in enumerate(header):
        column = header[col_index]
        steps_in_column = participant_data[column]
        if col_index > 3 and steps_in_column != '': # as the step data starts from 5th column, i.e. indexed 4
            total_steps = total_steps + int(steps_in_column)
    return total_steps

def get_team_names(steps_data):
    team_names = []
    for row in steps_data:
        if row['team_name'] not in team_names:
            team_names.append(row['team_name'])
    return team_names

def get_total_steps_by_team_name(steps_data, header, team_name):
    team_total_steps = 0
    for row in steps_data:
        if row['team_name'] == team_name:
            team_total_steps = team_total_steps + total_steps_per_person(row, header)
    return team_total_steps

with open('../../Biggest Loser 2018.csv') as csvfile: # IMPT: if you are running the code from the original location of Lesson 4, change '../../Biggest Loser 2018.csv' to 'Biggest Loser 2018.csv'
    readCSV = csv.DictReader(csvfile)
    # Get the header and pass to the function get_total_steps_by_team_name_from_data(steps_data, header, team_name)
    header = readCSV.fieldnames
    # For avoiding reading from a file multiple times(because it's not efficient), convert the data to a list for reuse
    stepsData = list(readCSV)
    team_names = get_team_names(stepsData)
    max_steps = -1 # it is used for comparison later, before the comparison starts, we want to have a number that is smaller than every team's total steps 
    winning_team = "Not found yet"
for name in team_names:
    total_steps_by_team = get_total_steps_by_team_name(stepsData, header, name)
    if total_steps_by_team > max_steps:
        # update the max_steps and winning_team if current team achieved more steps than the current winning_team
        max_steps = total_steps_by_team
        winning_team = name
print("The winning team is " + winning_team + " with " + str(max_steps) + " steps.")
````

<span>3. </span> (Very challenging) Following 2c, write code to find the first person who hit 400,000 steps.
Hints: 
* To read in the csv and store it as a table, which would be represented by an array of arrays.
* To update the table of steps to a table of accumulated steps.
eg: 
```
| 1 | 2 | 3 | 4 |       | 1 | 3 |  6 | 10 |
| 4 | 3 | 2 | 1 |  -->  | 4 | 7 |  9 | 10 |
| 3 | 4 | 5 | 6 |       | 3 | 7 | 12 | 18 |
```
* To loop through the table to find the first person who fulfills the condition.

````python
# Start challenging question 3 here.
import csv

def update_accumulated_steps(filtered_stepsData):
    num_row = len(filtered_stepsData)
    num_col = len(filtered_stepsData[0])
    for col_index in range(0, num_col): # loop through every column
        for row_index in range(0, num_row): # loop through every row
            current_day_steps = filtered_stepsData[row_index][col_index]
            
            if current_day_steps == "": # if the cell is empty, we convert if to zero
                current_day_steps = 0 
            else: # if the cell is not empty, we convert it from string to integer
                current_day_steps = int(current_day_steps)
            
            previous_day_steps = 0
            if col_index > 0: # if the column is not the first column, we will update it by adding the entry in front to the current cell
                previous_day_steps = filtered_stepsData[row_index][col_index - 1]
            
            filtered_stepsData[row_index][col_index] = previous_day_steps + current_day_steps
            
    return filtered_stepsData

def find_first_reach(accumulated_steps, cutoff_steps):
    num_row = len(accumulated_steps)
    num_col = len(accumulated_steps[0])
    for col_index in range(0, num_col): # loop through every column
        for row_index in range(0, num_row): # loop through every row
            if accumulated_steps[row_index][col_index] > cutoff_steps: # if find the cell with the accumulated steps greater than cutoff_steps
                print("The first to reach " + str(cutoff_steps) + " steps is Participant " + str(row_index + 1) + " on day " + str(col_index + 1) + ".")
                return 
    return

with open('../../Biggest Loser 2018.csv') as csvfile: # IMPT: if you are running the code from the original location of Lesson 4, change '../../Biggest Loser 2018.csv' to 'Biggest Loser 2018.csv'
    readCSV = csv.reader(csvfile)
    stepsData = list(readCSV)
    total_days = len(stepsData[0]) - 4
    filtered_stepsData = [] # create a table to save the steps information only
    for i in range(1, len(stepsData)): # to loop from the second line of the csv file
        filtered_stepsData.append(stepsData[i][4:]) # to loop from the 5th column onwards
    accumulated_steps = update_accumulated_steps(filtered_stepsData)
    find_first_reach(accumulated_steps, 400000)
````
